```python
def generate_children(s):
  l=[]
  for i in range(len(s)):
    if(len(s[i])!=0):
      for j in range(len(s)):
       s1=copy.deepcopy(s)
       if(j!=i):
        x=s1[i][0]
        s1[i].remove(x)
        s1[j].insert(0,x)
        l.append(s1)
  return l
```
